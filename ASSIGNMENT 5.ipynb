# Task 1:
1. Using OpenCV functions, open the webcam / USB camera device and play a live stream
2. Video stream must stop when "x" key is pressed<br>

```python
import cv2

# Create a VideoCapture object
cap = cv2.VideoCapture(0)

# Check if the webcam is opened successfully
if not cap.isOpened():
  print("Could not open webcam")
  exit()

# Create a window to display the video stream
cv2.namedWindow('Webcam')

# Infinite loop to read frames from the webcam
while True:
  # Read the frame from the webcam
  ret, frame = cap.read()

  # Display the frame in the window
  cv2.imshow('Webcam', frame)

  # Check if the "x" key is pressed
  if cv2.waitKey(1) == ord('x'):
    break

# Release the resources
cap.release()
cv2.destroyAllWindows()
```

# Task 3:
1. In the live camera feed, using the Haar Cascade Files, detect following:-
    - Face (Enclosed with Yellow Rectangle)
    - Nose (Enclosed with Blue Rectangle)
    - Eyes (Enclosed with Green Rectangle)
    - Smile (Enclosed with Red Rectangle)<br>
**Hint-** Refer Section 50

```python
import cv2

# Load the Haar Cascade classifiers
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
nose_cascade = cv2.CascadeClassifier('haarcascade_nose.xml')
eyes_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
smile_cascade = cv2.CascadeClassifier('haarcascade_smile.xml')

# Create a VideoCapture object
cap = cv2.VideoCapture(0)

# Check if the webcam is opened successfully
if not cap.isOpened():
    print("Could not open webcam")
    exit()

# Create two windows to display the video stream and grayscale feed
cv2.namedWindow('Webcam')
cv2.namedWindow('Grayscale')

# Infinite loop to read frames from the webcam
while True:
  # Read the frame from the webcam
  ret, frame = cap.read()

  # Convert the frame to grayscale
  gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

  # Detect faces in the grayscale frame
  faces = face_cascade.detectMultiScale(gray, 1.1, 5)

  # Draw a yellow rectangle around each face
  for (x, y, w, h) in faces:
    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 255), 2)

  # Detect noses in the grayscale frame
  noses = nose_cascade.detectMultiScale(gray, 1.1, 5)

  # Draw a blue rectangle around each nose
  for (x, y, w, h) in noses:
    cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

  # Detect eyes in the grayscale frame
  eyes = eyes_cascade.detectMultiScale(gray, 1.1, 5)

  # Draw a green rectangle around each eye
  for (x, y, w, h) in eyes:
    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

  # Detect smiles in the grayscale frame
  smiles = smile_cascade.detectMultiScale(gray, 1.1, 5)

  # Draw a red rectangle around each smile
  for (x, y, w, h) in smiles:
    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)

  # Display the frames
  cv2.imshow('Webcam', frame)
  cv2.imshow('Grayscale', gray)

  # Check if the "x" key is pressed
  if cv2.waitKey(1) == ord('x'):
    break

# Release the resources
cap.release()
cv2.destroyAllWindows()
```
